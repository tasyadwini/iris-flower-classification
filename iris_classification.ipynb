#Read Dataset

```python
import pandas as pd
from sklearn import datasets

# Making iris dataset and converted it into DataFrame
iris = datasets.load_iris()
x = iris.data
y = iris.target

# Convert feature data and target into Dataframe
df_x = pd.DataFrame(x, columns=iris.feature_names)
df_y = pd.Series(y, name='target')

# Merge feature and target into one Dataframe
df = pd.concat([df_x, df_y], axis=1)

df.head(10)
```

```python
df.info()
```

```python
df['target'].unique()
```

#Split Dataset

```python
from sklearn.model_selection import train_test_split

# Splitting data into train dan test
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

#Train The Model

```python
from sklearn.tree import DecisionTreeClassifier

# Making and training the Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

#Predict and Evaluate

```python
from sklearn.metrics import accuracy_score

# Predict and Evaluate
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print("Classification Report: ")
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Visualisasi decision tree (opsional)
plt.figure(figsize=(20,10))
tree.plot_tree(model,
               feature_names=iris.feature_names,
               class_names=iris.target_names, filled=True)
plt.show()
```
